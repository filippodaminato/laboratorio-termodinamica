# ESPERIENZA 0

```python
df = pd.read_csv("alto_basso.csv")
# s K°

t = df["Tempo"].to_numpy()
T = df["Temperatura"].to_numpy()

plt.plot(t,T)
```

```python
T_log = []
for e in T:
    T_log.append((290 - e)/(290 - 315))

uy= np.repeat(0.1,len(t))

plt.plot(t,np.log(T_log),linestyle="",marker=".")
```

```python
m0, sm0, c0, sc0, cov0, rho0 = lib.lin_fit(t,np.log(T_log), uy, "Tempo", "Temperatura", 22, 32, -3.5, 0, plot=True, setrange=False)
```

# ESPERIENZA A

```python
df = pd.read_csv("posizione_a_2.txt",sep="\t")
# s K°
df['Tempo'] = [float(str(i).replace(",", ".")) for i in df["Tempo"]]
df["Posizione"] = [float(str(i).replace(",", ".")) for i in df["Posizione"]]

Tempo = np.array(df["Tempo"])
Posizione = np.array(df["Posizione"])
```

```python
x = pd.read_csv("pressione_a_2.txt",sep="\t")
# s K°
x['Tempo'] = [float(str(i).replace(",", ".")) for i in x["Tempo"]]
x["Pressione"] = [float(str(i).replace(",", ".")) for i in x["Pressione"]]

Tempo = np.array(x["Tempo"])
Pressione = np.array(x["Pressione"])
```

```python
plt.plot(Tempo,Pressione)
plt.xlabel("Tempo (s)")
plt.ylabel("Pressione (kPa)")
```

```python
plt.plot(Tempo,Posizione)
plt.xlabel("Tempo (s)")
plt.ylabel("Posizione (m)")
```

```python
plt.plot(Pressione,Posizione)
plt.xlabel("Pressione (kPa)")
plt.ylabel("Posizione (m)")
```

```python
delta_x = Posizione[(Tempo<22.72) & (Tempo>22.55)]
delta_p = Pressione[(Tempo<22.72) & (Tempo>22.55)]

plt.plot(delta_p,delta_x)
plt.xlabel("Pressione (kPa)")
plt.ylabel("Posizione (m)")
```

```python
diametro = 0.0325
area = (diametro*diametro*math.pi)/4
print(area)
print(math.pi)
```

```python
Volume = delta_x*area*1000000
uy = np.repeat(0.1,len(Volume))
```

```python
plt.plot(delta_p,Volume)
plt.xlabel("Pressione (kPa)")
plt.ylabel("Volume (cm^3)")
```

```python
m0, sm0, c0, sc0, cov0, rho0 = lib.lin_fit(delta_p,Volume, uy, "Pressione", "Volume", 22, 32, -3.5, 0, plot=True, setrange=False)
```
