```python
# plots will be shown inline 
%matplotlib inline 
import matplotlib.pyplot as plt

import numpy
from numpy import sqrt,floor

import numpy as np
import scipy as sp
from scipy import stats

import pandas as pd
import random
import math

# libreria locale
import my_lib_santanastasio as lib

m = [53.9,69.3,84.6] #gr
base_area = (np.pi * pow(3.25/2, 2)) #cm^3

v0 = base_area * 2 #cm^3
```

```python
df = pd.read_csv("dati/pp1.txt",delim_whitespace=True,decimal=',')
# s K°

pos = df["Posizione"].to_numpy() #m
pres = df["Pressione"].to_numpy() #kPa

points = [
    [ 7059, 'P4' ],
    [ 6555, 'P3' ],
    [ 2584,'P2' ],
    [ 0, 'P1' ],
]

vol = pos * 100 * base_area  

plt.plot(vol[::50], pres[::50], marker='o',linestyle="dashed")

for i in points:
    plt.annotate(i[1], (vol[i[0]], pres[i[0]]), color="black")
    plt.scatter(vol[i[0]], pres[i[0]], color="red",zorder=10)
```
